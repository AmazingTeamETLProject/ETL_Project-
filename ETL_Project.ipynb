```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
fertility_rate = "fertility_rate.csv"
fertility_df = pd.read_csv(fertility_rate)
fertility_df.head()
```

```python
fertility_new_columns = fertility_df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
fertility_new_columns.index.name = 'id'
fertility_new_columns
```

```python
fertility_new_columns = fertility_new_columns.rename(columns={'Country Name': 'country_name'})
fertility_new_columns
```

```python
life_expectancy = "life_expectancy.csv"
life_expectancy_df = pd.read_csv(life_expectancy)
life_expectancy_df.head()
```

```python
life_expectancy_new_columns = life_expectancy_df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
life_expectancy_new_columns.index.name = 'id'
life_expectancy_new_columns
```

```python
life_expectancy_new_columns = life_expectancy_new_columns.rename(columns={'Country Name': 'country_name'})
life_expectancy_new_columns
```

# Create Database Connection

```python
connection_string = "root:Brezel02!@localhost/world_etl_db"
engine = create_engine(f'mysql+pymysql://{connection_string}')
```

```python
# find all tables in world_etl_db database in SQL
engine.table_names()
```

```python
# transfer pandas dataframe to life_expectancy table, use the replace to automatically fill
# in all columns with data
life_expectancy_new_columns.to_sql(name='life_expectancy', con=engine, if_exists='replace', index=True)

```

```python
# transfer pandas dataframe to fertility_rate table, use the replace to automatically fill
# in all columns with data
fertility_new_columns.to_sql(name='fertility_rate', con=engine, if_exists='replace', index=True)

```

```python
# how to aggregate across rows:
pd.read_sql_query('''SELECT (life_expectancy.1970 + life_expectancy.1980 + life_expectancy.2000)/3 AS Average
FROM life_expectancy
WHERE country_name = 'Italy';''', con=engine).head()
```
